```python
'''
[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"},
{"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] =>
['Action','Adventure','Fantasy','SciFi']


it is a string of list so lets convert it using ast module
'''
import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
# for keywords we could do the same

movies['keywords']=movies['keywords'].apply(convert)
```

```python
'''
lets do cast for cast lets take only top 3 actors
lets reuse convert function
'''

def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
'''
lets take the director name from crew
'''

def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import ast

def clean_list_column(x):
    # Convert string representation of list to actual list (if needed)
    if isinstance(x, str):
        try:
            x = ast.literal_eval(x)
        except:
            return []
    # If not a list, skip
    if not isinstance(x, list):
        return []
    return [i.replace(" ", "") for i in x if isinstance(i, str)]

movies['genres'] = movies['genres'].apply(clean_list_column)
movies['keywords'] = movies['keywords'].apply(clean_list_column)
movies['cast'] = movies['cast'].apply(clean_list_column)
movies['crew'] = movies['crew'].apply(clean_list_column)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies['tags']=movies['tags'].apply(lambda x: x.lower())
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
'''
instead of euclidian distance we will use cosine distance
the moer the angle the lesser the similarity
'''

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("The Dark Knight Rises")
```

```python
recommend('The Dark Knight')
```

```python
recommend('Avatar')
```
